```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
url = "https://libris-qa.kb.se/find?q=inträdestal%20i%20Svenska%20akademien&_limit=1000"
#url = "https://libris-qa.kb.se/find?q=Svenska%20akademien&limit=1000"
#r = http.request('GET', url) 
r = http.request('GET', url, headers={'Accept': 'application/json'})
#r = http.request('GET', url, headers={'Accept': 'application/ld+json'})
#data = json.loads(r.data.decode('utf-8')) 
print('Status: ',r.status)
#print(r.info())  
#print(type(r.data))
#print(r.data)  

data = json.loads(r.data.decode('utf-8')) 
#print(data)  
print(len(data["items"]))  

for row in (data["items"]):
    #print("id", row["@id"])
    #https://libris-qa.kb.se/s4xhfv2jqxbmww72#it
    #https://libris-qa.kb.se/s4xhfv2jqxbmww72/data.jsonld
    url_instance = row["@id"].replace("#it","/data.jsonld")
    print("url_instance: ",url_instance)
    print("\tjsonld", row["@id"].replace("#it","/data.jsonld"))
    try:
        print("\tTitle:", row["hasTitle"][0]['mainTitle'])
    except:
        pass
    try:
        print("\tSubtitle:", row["hasTitle"][0]['subtitle'])
    except:
        pass
    try:
        print("\tidentifiedBy type:", row["identifiedBy"]["@type"])
    except:
        pass
    try:
#        print("\tInstanceOf:", row["instanceOf"]["@id"])
#        print("Summary 2:", row["instanceOf"]["subject"])
        print("\tType Instance:", type(row["instanceOf"]))
        print("\tKeys:", row["instanceOf"].keys())
        print("\tSummary:")
        print("\tOld LIBRIS:")
        print("\tPublication year:")

    except:
        pass
    

```
