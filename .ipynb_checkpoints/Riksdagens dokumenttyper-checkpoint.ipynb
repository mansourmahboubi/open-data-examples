```python
import urllib3, json
import pandas as pd  
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
pd.set_option("display.max.rows", None) 
url ="http://data.riksdagen.se/sv/koder/?typ=doktyp&utformat=json"

r = http.request('GET', url)
data = json.loads(r.data)
dftot = pd.DataFrame(data["typer"]["typ"])
dftot.head()
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt   
plot = dftot.familj.value_counts().plot.bar(y='counts', figsize=(25, 5)) 
plt.show()
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt   
plot = dftot.familj.value_counts().plot.pie(y='counts', figsize=(25, 5)) 
plt.show()
```
