* intressent_id = Wikidata Property [P1214](https://www.wikidata.org/wiki/Property:P1214) 
* sourceid = Wikidata Property [P8388](https://www.wikidata.org/wiki/Property:P8388)

 * Tjänstgörande talmansersättare
   



```python
from datetime import datetime
start_time  = datetime.now()
print("Last run: ", start_time)
```

```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
url ="http://data.riksdagen.se/personlista/?rdlstatus=samtliga&utformat=json&termlista="

r = http.request('GET', url)
data = json.loads(r.data)
dftot = pd.DataFrame(data["personlista"]["person"])
dftot.head()
```

```python
dftot.info()
```

```python
dftot[['status','sourceid','tilltalsnamn','efternamn','parti','valkrets']][0:50]
```

```python
dftot.status.value_counts()[1:50]
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt   
plot = dftot.status.value_counts()[0:20].plot.bar(y='counts', figsize=(25, 5)) 
plt.show()
```

#### Tjänstgörande 
borde vara 349

```python

dfnorm = pd.json_normalize(data["personlista"]["person"])
```

```python
#dfnorm
dfnormTj = dfnorm[(dfnorm["status"] == "Tjänstgörande riksdagsledamot") | \
                   (dfnorm["status"] == "Tjänstgörande statsrådsersättare") | \
                   (dfnorm["status"] == "Tjänstgörande ersättare") | \
                   (dfnorm["status"] == "Andre vice talman (tjänstgörande riksdagsledamot)") | \
                   (dfnorm["status"] == "Tjänstgörande talmansersättare") ]
   
dfnormTj.info()
```

```python
dfnormTj.status.value_counts()
```
