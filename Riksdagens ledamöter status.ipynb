```python
from datetime import datetime
start_time  = datetime.now()
print("Last run: ", start_time)
```

```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
url ="http://data.riksdagen.se/personlista/?rdlstatus=samtliga&utformat=json&termlista="

r = http.request('GET', url)
data = json.loads(r.data)
dftot = pd.DataFrame(data["personlista"]["person"])
dftot.head()
```

#### Tjänstgörande 
borde vara 349

```python

dfnorm = pd.json_normalize(data["personlista"]["person"])
```

```python
#dfnorm
dfnormTj = dfnorm[(dfnorm["status"] == "Tjänstgörande riksdagsledamot") | \
                   (dfnorm["status"] == "Tjänstgörande statsrådsersättare") | \
                   (dfnorm["status"] == "Tjänstgörande ersättare") | \
                   (dfnorm["status"] == "Förste vice talman (tjänstgörande riksdagsledamot)") | \
                   (dfnorm["status"] == "Andre vice talman (tjänstgörande riksdagsledamot)") | \
                   (dfnorm["status"] == "Tredje vice talman (tjänstgörande riksdagsledamot)") | \
                   (dfnorm["status"] == "Tjänstgörande talmansersättare") ]
   
dfnormTj.info()
```

```python
dfnormTj.status.value_counts()
```

```python
pd.set_option('display.max_rows', None)

# Define the sorter
sorter = ['Förste vice talman (tjänstgörande riksdagsledamot)','Andre vice talman (tjänstgörande riksdagsledamot)',\
         'Tredje vice talman (tjänstgörande riksdagsledamot)','Tjänstgörande talmansersättare',\
         'Tjänstgörande ersättare ','Tjänstgörande statsrådsersättare','Tjänstgörande riksdagsledamot']
sorterIndex = dict(zip(sorter, range(len(sorter))))
dfnormTj['Status_Rank'] = dfnormTj['status'].map(sorterIndex)
dfnormTj = dfnormTj.sort_values(by=['Status_Rank','parti','efternamn','tilltalsnamn'])

# get index start 1
dfnormTj = dfnormTj.reset_index(drop=True)
dfnormTj.index += 1 
dfnormTj[['status','parti','efternamn','tilltalsnamn','valkrets','sourceid']]
```

```python
plotTj = dfnormTj.status.value_counts()[0:20].plot.bar(y='counts', figsize=(25, 5)) 
plt.show()
```

```python
dfnormTju = dfnormTj.groupby(['parti']).status.value_counts().unstack()
dfnormTju = dfnormTju.fillna(0)
pd.option_context('display.precision', 0)

s= dfnormTju.style.format('{:.0f}') 
s
```

```python
plotTju = dfnormTju.plot.bar( figsize=(25, 5)) 
plt.show()
```

```python
#Ersättare
dfnormTjErs = dfnormTj[dfnormTj["status"] == "Tjänstgörande ersättare"]
dfnormTjErs
```
