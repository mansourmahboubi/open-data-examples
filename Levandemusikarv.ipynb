```python
# pip install sparqlwrapper
# https://rdflib.github.io/sparqlwrapper/

import sys
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd 
from bs4 import BeautifulSoup
import sys,re
import urllib3
from tqdm.notebook import trange  
endpoint_url = "https://query.wikidata.org/sparql"
urlbase = "http://www.levandemusikarv.se"

query = """SELECT ?item  ?Levid ?url WHERE {
  ?item wdt:P4607 ?Levid.
  BIND (URI(CONCAT("http://www.levandemusikarv.se/tonsattare/",?Levid)) AS ?url )
}"""



def get_results(endpoint_url, query):
    user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
    # TODO adjust user agent; see https://w.wiki/CX6
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()

http = urllib3.PoolManager()

def getCompositions(url):
    listVerk = {}
    r = http.request('GET', url) 
    soup = BeautifulSoup(r.data, "html.parser")
    global comp
    for link in soup.select("a[href*=SMH-W]"):
        comp += 1

        url_comp = urlbase + link['href']
        print(url_comp)
        r_comp = http.request('GET', url_comp) 
        soup_comp = BeautifulSoup(r_comp.data, "html.parser")
        #print(soup_comp)
        for li_tag in soup_comp.find_all('li'):
            #print("li_tag.text:",li_tag.text)

            #if "verktyp:" in li_tag.text:
            if "Verktyp:" in li_tag.text:
                print("\t", li_tag.text)
            if "Textförfattare:" in li_tag.text:
                print("\t", li_tag.text)
            if "Uruppförande:" in li_tag.text:
                print("\t", li_tag.text)
            if "Speltid:" in li_tag.text:
                print("\t", li_tag.text)
            if "Tillkomstår:" in li_tag.text:
                print("\t", li_tag.text)
    return listVerk
        

results = get_results(endpoint_url, query)
comp = 0
for result in results["results"]["bindings"]:
    wd = result["item"]["value"].replace("http://www.wikidata.org/entity/","")
    url = result["url"]["value"]
    print(wd, url)
    getCompositions(url)
```

```python
print ("\n\nNumber of compositions found: ", comp)
```
