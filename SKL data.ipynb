### Kommunkoder

```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
 
url = "https://catalog.skl.se/rowstore/dataset/8621fe21-0120-407e-a81f-705ef45f76d2" 
header = {'Accept': 'application/json'}
r = http.request('GET', url)
#print(r.data)
data = json.loads(r.data)
df = pd.DataFrame(data["results"])
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt 
df["folkmängd 31 december 2018"].str.replace(" ","").astype(int).plot.bar() 
plt.show()
```
