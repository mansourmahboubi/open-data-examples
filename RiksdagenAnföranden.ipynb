## Riksdagen Anföranden
* [Riksdagensdata anföranden](http://data.riksdagen.se/anforandelista/?rm=&anftyp=&d=&ts=&parti=&iid=&sz=10000&utformat=json)

```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
 
url = "http://data.riksdagen.se/anforandelista/?rm=&anftyp=&d=&ts=&parti=&iid=&sz=1000&utformat=json"
r = http.request('GET', url)
data = json.loads(r.data)
```

```python
df = pd.DataFrame(pd.DataFrame(data["anforandelista"]["anforande"]))
df
```
