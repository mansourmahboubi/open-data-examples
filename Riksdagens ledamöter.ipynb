## Check Wikidata items and get new GUID
SPARQL: https://w.wiki/VdW 

1. finds item with short id and no GUID 
1. Get JSON from PM and extract GUID
1. Upload in batch using Quickstatement see [Batch 37989](https://quickstatements.toolforge.org/#/batch/37989)

```python
# pip install sparqlwrapper
# https://rdflib.github.io/sparqlwrapper/

import sys,urllib3,json
from SPARQLWrapper import SPARQLWrapper, JSON
http = urllib3.PoolManager() 
endpoint_url = "https://query.wikidata.org/sparql"

query = """SELECT ?item ?old  WHERE {
?item wdt:P1214 ?old.
FILTER (STRLEN(?old) < 15)
minus {?item wdt:P8388 ?guid}
}"""

# 
def get_results(endpoint_url, query):
    user_agent = "user salgo60/%s.%s" % (sys.version_info[0], sys.version_info[1])
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()

def get_GUID(intressent_id):
    baseURL = "http://data.riksdagen.se/personlista/?utformat=json&iid="
    url = baseURL + intressent_id
    #print(url)
    try:
        r_MP = http.request('GET', url) 
        dataMP = json.loads(r_MP.data) 
        guid = dataMP["personlista"]["person"]["sourceid"]
        return guid
    except Exception as e:
        print(e)
    return False
    
results = get_results(endpoint_url, query)

for result in results["results"]["bindings"]:
    wd = result["item"]["value"].replace("http://www.wikidata.org/entity/","")

    intressent_id = result["old"]["value"]
    guid = get_GUID(intressent_id)
    print(wd,";",intressent_id,";",guid)
```
