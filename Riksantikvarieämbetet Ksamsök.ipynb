```python
def checkURL(externalurl):
    r = http.request('GET', externalurl)
    if r.status != 200:
        print("Status: ",r.status, " \t", externalurl)
        return False
    return True
```

```python
import urllib3, json
import pandas as pd   
from bs4 import BeautifulSoup
from tqdm.notebook import trange 
  
http = urllib3.PoolManager()
urlbase = "https://sv.wikipedia.org/w/index.php?title=Special:L%C3%A4nks%C3%B6kning&limit=5000&target=http%3A%2F%2Fkulturarvsdata.se&offset="

ok = 0
notok = 0
dftot = pd.DataFrame()
for i in trange(0,60000,5000):
    url = urlbase + str(i)
    r = http.request('GET', url) 
    soup = BeautifulSoup(r.data, "html.parser")
    for t in range(1, len(soup.findAll('a'))):
        one_a_tag = soup.findAll('a')[t]
        one_a_tag_href = one_a_tag['href']
        try:
            if one_a_tag['class'][0] == "external":
                if checkURL(one_a_tag_href):
                    ok += 1
                else:
                    notok += 1
        except:
            pass
print("OK: ",ok,"\t not ok",notok)
```
