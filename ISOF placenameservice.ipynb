```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
urlbase = "https://placenameservice.isof.se/PlaceNameService/" 
url_service = "v1/geo-reference/parishes"

url = urlbase + url_service + "?format=json"
data_parishes = pd.read_json(url)
data_parishes.info()    
```
