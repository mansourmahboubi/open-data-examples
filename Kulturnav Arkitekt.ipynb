```python
from datetime import datetime
start_time  = datetime.now()

print("Last run: ", start_time)
```

```python
import urllib3, json
import pandas as pd  
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
urlbase ="http://kulturnav.org/api/list/2b7670e1-b44e-4064-817d-27834b03067c"
#urlbase = "https://kulturnav.org/api/search/entity.sameAs:https%3A%2F%2Fwww.wikidata.org%2Fentity%2FQ*/" 
try:
    r = http.request('GET', urlbase)
    data = json.loads(r.data)
except Exception as e: print(e)
```

```python
for d in data:
    wikidata = ""
    try:
        for s in d["properties"]["entity.sameAs"]:
            if "wikidata" in s["value"]:
                wikidata = s["value"]
    except:
        pass    
    print(d["uuid"],"|",d["entityType"],"|",d["caption"]["sv"],"|",wikidata )
```
