### Exempel "Särskild beräkning av tandvårdsersättning med anledning av sjukdomen covid-19" - Socialutskottets bet 2019/20:SoU26 - H701SoU26
* [html www.riksdagen.se](https://www.riksdagen.se/sv/dokument-lagar/arende/betankande/_H701SoU26) [text](http://data.riksdagen.se/dokument/H701SoU26.text) / [html data](http://data.riksdagen.se/dokument/H701SoU26.html) /  [json](http://data.riksdagen.se/dokumentlista/?sok=H701SoU26&utformat=json) 


```python
from datetime import datetime
start_time  = datetime.now()
print("Last run: ", start_time)
```

```python
import urllib3, json
import pandas as pd  
from tqdm.notebook import  trange   
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
pd.set_option("display.max.rows", None) 
urlbase ="http://data.riksdagen.se/dokumentlista/?sok=&doktyp=bet&utformat=json&a=s&start="

dftot = pd.DataFrame()
maxpages = int(69138/20) + 1
for i in trange(1,maxpages):
    url = urlbase + str(i)
    r = http.request('GET', url)
    try:
        data = json.loads(r.data)
        dftot = dftot.append(pd.DataFrame(data["dokumentlista"]["dokument"]),sort=False)
    except:
        pass
        print("Error : " + url)

```
