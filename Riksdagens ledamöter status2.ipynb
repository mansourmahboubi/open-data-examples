```python
from datetime import datetime
start_time  = datetime.now()
print("Last run: ", start_time)
```

```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 
url ="http://data.riksdagen.se/personlista/?iid=&fnamn=&enamn=&f_ar=&kn=&parti=&valkrets=&org=&utformat=json"

r = http.request('GET', url)
data = json.loads(r.data)
dftot = pd.DataFrame(data["personlista"]["person"])
dftot.head()
```

```python
dftot.info()
```

```python
dftot.status.value_counts()
```

```python
pd.set_option('display.max_rows', None)
# Define the sorter
sorter = ['Förste vice talman (tjänstgörande riksdagsledamot)','Andre vice talman (tjänstgörande riksdagsledamot)',\
         'Tredje vice talman (tjänstgörande riksdagsledamot)','Tjänstgörande talmansersättare',\
         'Tjänstgörande ersättare ','Tjänstgörande statsrådsersättare','Tjänstgörande riksdagsledamot']
sorterIndex = dict(zip(sorter, range(len(sorter))))
dftot['Status_Rank'] = dftot['status'].map(sorterIndex)
dftot = dftot.sort_values(by=['Status_Rank','parti','efternamn','tilltalsnamn'])

# get index start 1
dftot = dftot.reset_index(drop=True)
dftot.index += 1 
dftot[['status','sourceid','tilltalsnamn','efternamn','parti','valkrets']]
```

```python

dfnormTju = dftot.groupby(['parti']).status.value_counts().unstack()
dfnormTju = dfnormTju.fillna(0)
pd.option_context('display.precision', 0)

s= dfnormTju.style.format('{:.0f}') 
s
```

```python
plotTju = dfnormTju.plot.bar( figsize=(25, 5)) 
```
